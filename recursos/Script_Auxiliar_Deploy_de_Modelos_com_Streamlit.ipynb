```python
from pycaret.datasets import get_data
dados = get_data('heart_disease')
```

```python
dados.describe()
```

```python
from pycaret.regression import *
```

```python
#remover espaço em branco do nome da coluna
dados.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
class_setup = setup(data = dados,
                    target = 'Disease',
                    numeric_features = ['age','resting blood pressure', 'serum cholestoral in mg/dl', 'maximum heart rate achieved', 'oldpeak'],
                    categorical_features = ['sex', 'chest pain type', 'fasting blood sugar > 120 mg/dl', 'resting electrocardiographic results', 'exercise induced angina', 'slope of peak', 'number of major vessels', 'thal'],
                    normalize = True,
                    normalize_method = 'minmax',
                    n_jobs = -1)
```

```python
# class_setup = setup(data = dados,
#                     target = 'charges',
#                     normalize = True,
#                     n_jobs = -1)
```

```python
modelo = compare_models()
```

```python
modelo_tunado = tune_model(modelo, optimize = 'RMSE')
```

```python
modelo_final = finalize_model(modelo_tunado)
```

```python
save_model(modelo_final, 'modelo-previsao-doenca-coracao')
```
