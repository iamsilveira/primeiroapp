# Formulação de Métrica de Custo:

Para formular a métrica de custo, onde o custo menor será o melhor caso, levei em consideração uma pontuação que favoreça o verdadeiro negativo (custo 0), ou seja, previu que não era doença e não era doença. Em seguida, favoreça o verdadeiro positivo, ou seja, previu que era doença e era doença (custo 1).

Em contrapartida, levei em consideração uma pontuação que desfavoreça mais o falso negativo (custo 4), ou seja, previu não doença e estava doente. Em seguida, desfavoreça o falso positivo (custo 2), ou seja, previu doença e não era doença.

Pelo entendimento, é mais grave o falso negativo, por ter implicações mais sérias tanto para o paciente quanto para a seguradora. Por esse motivo, o seu custo tem um peso maior, impactando diretamente na escolha do melhor modelo.

Os modelos serão analisados levando em consideração o menor custo possível.

```python
# Criando uma métrica personalizada no PyCaret
def custo(y, y_pred):
    vp = np.where((y_pred==1) & (y==1), 1, 0)  # 1 previu doença e era doença -> custo do tratamento
    fp = np.where((y_pred==1) & (y==0), 2, 0)  # 2 previu doença e não era doença -> custo do tratamento, mas não era necessário
    vn = np.where((y_pred==0) & (y==0), 0, 0)  # 0 previu não doença e não era doença -> custo 0
    fn = np.where((y_pred==0) & (y==1), 4, 0)  # 4 previu não doença, mas estava doente -> pior caso do custo, custo do tratamento tardio e possível indenização judicial
    return np.sum([vp, fp, vn, fn])/len(y)
```

```python
# add metric to PyCaret
#remove_metric('custo')
add_metric('custo', 'Custo', custo, greater_is_better = False)
```

```python
# CRIAÇÃO DE UMA BASE PARA VALIDAÇÃO

aux = dados.sample(frac=1)
teste = aux.iloc[:81]
treino = aux.iloc[81:]

treino.reset_index(drop = True, inplace = True)
teste.reset_index(drop = True, inplace = True)

print(treino.shape)
print(teste.shape)
```

##  UNDERSAMPLING

```python
# 1. SEPARAR A BASE EM DUAS, DE ACORDO COM A VARIAVEL RESPOSTA
dados0 = treino[treino['Disease'] == 0]
dados1 = treino[treino['Disease'] == 1]
print(dados0.shape)
print(dados1.shape)
```

```python
# 2. RETIRAR UMA AMOSTRA ALEATORIA DA BASE MAIOR, DO TAMANHO DA BASE MENOR
amostra0 = dados0.sample(81)

# 3. JUNTAR A AMOSTRA COM A BASE E FORMAR O CONJUNTO DE DADOS PARA MODELAGEM
dados_under = pd.concat([amostra0, dados1])
dados_under.shape
```

## OVERSAMPLING

```python
maior = pd.concat(2*[dados1])
amostra1 = maior.sample(108)

dados_over = pd.concat([dados0, amostra1])
dados_over.reset_index(drop = True, inplace = True)
dados_over.shape
```
